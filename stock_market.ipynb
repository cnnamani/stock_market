```python

import pandas as pd
from yahoo_fin import stock_info
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

def My_portfolio1():
    df = pd.DataFrame({
        'stock_names':['NIO','JMIA','SVRA','SMURF'],
        'quantity':[16,50,796,335],
        'entry_price':[ 23.76,9.86,1.10,1.15]






    }

    )

# getiing the real time stock price
    N= stock_info.get_live_price('NIO')
    J = stock_info.get_live_price('JMIA')
    S = stock_info.get_live_price('SVRA')
    SM = stock_info.get_live_price('SMURF')
    
# inserting our real time stock price data in right rows and columns
    df['live_update'] = ''
    df.loc[0,'live_update']= N
    df.loc[1, 'live_update'] = J
    df.loc[2, 'live_update'] = S
    df.loc[3, 'live_update'] = SM
    
# creating a new data series that provides us with  useful information 
    df['book_value'] = df['quantity']  * df['entry_price']
    df['market_value']= df['quantity'] * df['live_update']
    df['market_value'] = df['quantity'] * df['live_update']
    df['market_value'] = df['quantity'] * df['live_update']
    df['market_value'] = df['quantity'] * df['live_update']

    df['profit_loss'] = df['market_value'] - df['book_value']


    df.plot()


    return df

    
print(My_portfolio1())
```
